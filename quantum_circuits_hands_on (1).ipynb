# 1. What is a bit in classical computing and a qubit in quantum computing

A classical bit is either:

```text
0 or 1
```

A qubit can be:

```text
|0>, |1>, or a superposition of both
```

A general single-qubit state is written as:

$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle
$$

where:

$$
|\alpha|^2 + |\beta|^2 = 1
$$

Simple meaning:

- $|\alpha|^2$ is the probability of measuring `0`.
- $|\beta|^2$ is the probability of measuring `1`.

# 2. what is a Bloch sphere

The Bloch sphere is a 3D visualization for **one qubit**.

```text
Top of sphere    -> |0>
Bottom of sphere -> |1>
Equator          -> superposition states
```

please remember:

- Bloch sphere is excellent for visualisation but for single gates only
- For entangled two-qubit states, Bloch spheres alone do not show the full story. For entanglement we also use **statevector** and **measurement histogram**.

# 3. Initial state: |0>

qubits start in state $|0\rangle$ by default. As per the lib we are using

see below:

# 4. Initial state: |1>

But, we can create $|1\rangle$ by applying the `X` gate to $|0\rangle$.

## Pauli-X gate

The `X` gate is like a classical NOT gate.

$$
X|0\rangle = |1\rangle
$$

$$
X|1\rangle = |0\rangle
$$

````python
qc = QuantumCircuit(1)

qc.x(0)

display(qc.draw(output="mpl"))

state = Statevector.from_instruction(qc)

display(state.draw("latex"))
display(plot_bloch_multivector(state))
````

### Persistent Path code

````python
from qiskit.visualization.bloch import Bloch

num_steps = 40
angles = np.linspace(0, np.pi, num_steps)

xs = []
ys = []
zs = []

for theta in angles:
    qc = QuantumCircuit(1)

    # X gate shown slowly as Rx rotation
    qc.rx(theta, 0)

    state = Statevector.from_instruction(qc)

    a = state.data[0]
    b_amp = state.data[1]

    x = 2 * np.real(np.conj(a) * b_amp)
    y = 2 * np.imag(np.conj(a) * b_amp)
    z = np.abs(a)**2 - np.abs(b_amp)**2

    xs.append(x)
    ys.append(y)
    zs.append(z)

# Final X gate circuit
qc = QuantumCircuit(1)
qc.x(0)

# display(qc.draw(output="mpl"))

# state = Statevector.from_instruction(qc)

# display(state.draw("latex"))

# Bloch sphere with persistent path
bloch = Bloch()

bloch.point_color = ["r"]
bloch.vector_color = ["b"]

# red path
bloch.add_points([xs, ys, zs], meth="o")

# final blue arrow
bloch.add_vectors([xs[-1], ys[-1], zs[-1]])

bloch.view = [135, 30]

bloch.show()
````

### Measure after X gate

Since the final state is $|1\rangle$, measurement should give mostly `1`.

````python
qc = QuantumCircuit(1)

qc.x(0)
qc.measure_all()
display(qc.draw(output="mpl"))
## till now we have only added a measurment box, and not ran or did actual mesearements!

## now we run this circuit, this how we run any circuit:
simulator = AerSimulator()
# Run this quantum circuit qc on the simulator 1024 times. Because quantum measurement is probabilistic.
# One run gives one answer. Many runs give statistics. shots tells how many times to repeat the experiment.
result = simulator.run(qc, shots=1024).result()
counts = result.get_counts()

print(counts)
# display(plot_histogram(counts))
````

## Pauli-Y gate

The `Y` gate also moves $|0\rangle$ to $|1\rangle$, but with an extra phase factor.

$$
Y|0\rangle = i|1\rangle
$$

The measurement is still `1`, but the state has a phase.

````python
qc = QuantumCircuit(1)

qc.y(0)

display(qc.draw(output="mpl"))
state = Statevector.from_instruction(qc)
display(state.draw("latex"))
display(plot_bloch_multivector(state))

## the phase factor:
# print the amplitudes
amp_0 = state.data[0]
amp_1 = state.data[1]

print("Amplitude of |0> =", amp_0)
print("Amplitude of |1> =", amp_1)

# Phase of |1>
phase_rad = np.angle(amp_1)
phase_deg = np.degrees(phase_rad)

print("Phase angle of |1> =", phase_rad, "radians")
print("Phase angle of |1> =", phase_deg, "degrees")

print("Phase factor = i")
print("Y|0> = i|1>")

'''
This phase factor i does not change measurement result.
So measuring Y|0> still gives 1 always.
'''
````

## Pauli-Z gate

The `Z` gate does not change $|0\rangle$, but it changes the phase of $|1\rangle$.

$$
Z|0\rangle = |0\rangle
$$

$$
Z|1\rangle = -|1\rangle
$$

To see its effect clearly on the Bloch sphere, start from $|+\rangle$.

````python
# First it starts with |0⟩ by default
qc = QuantumCircuit(1)

# makes superposition:
# first make |+>
# |+⟩ = (|0⟩ + |1⟩) / √2
qc.h(0)
state = Statevector.from_instruction(qc)
display(state.draw("latex"))
display(plot_bloch_multivector(state))

# now apply Z
# qc.rz(theta, 0) // for custom angle, default is pi/2
qc.z(0)
'''
Z gate does this:

Z|0⟩ = |0⟩
Z|1⟩ = -|1⟩

So:

(|0⟩ + |1⟩) / √2

becomes:

(|0⟩ - |1⟩) / √2

This final state is called:

|-⟩

'''

display(qc.draw(output="mpl"))

state = Statevector.from_instruction(qc)
display(state.draw("latex"))
display(plot_bloch_multivector(state))


# Why we used H before Z?
# If we apply Z directly on |0⟩, nothing visible happens.
# So we first create |+⟩ using H, then Z shows its phase effect clearly.
````

### Persistent Path code

````python
from qiskit.visualization.bloch import Bloch

num_steps = 20
angles = np.linspace(0, np.pi, num_steps)

xs = []
ys = []
zs = []

for theta in angles:
    qc = QuantumCircuit(1)

    # First create |+>
    qc.h(0)

    # in loop apply Z as Rz(theta), for vis purpose
    qc.rz(theta, 0)

    state = Statevector.from_instruction(qc)

    a = state.data[0]
    b_amp = state.data[1]

    x = 2 * np.real(np.conj(a) * b_amp)
    y = 2 * np.imag(np.conj(a) * b_amp)
    z = np.abs(a)**2 - np.abs(b_amp)**2

    xs.append(x)
    ys.append(y)
    zs.append(z)

# Final Z gate circuit
qc = QuantumCircuit(1)
qc.h(0)
qc.z(0)

# display(qc.draw(output="mpl"))

state = Statevector.from_instruction(qc)

# display(state.draw("latex"))

# print("Amplitude of |0> =", state.data[0])
# print("Amplitude of |1> =", state.data[1])
# print("Final state = (|0> - |1>) / sqrt(2)")

bloch = Bloch()

bloch.point_color = ["r"]
bloch.point_marker = ["o"]
bloch.point_size = [30]
bloch.vector_color = ["b"]

# red dots
bloch.add_points([xs, ys, zs])

bloch.add_vectors([xs[-1], ys[-1], zs[-1]])

bloch.view = [130, 30]

bloch.show()
````

### Measurement

Measurement converts quantum information into classical output.

For one qubit:

```text
Possible outputs: 0 or 1
```

For two qubits:

```text
Possible outputs: 00, 01, 10, 11
```

Measurement is probabilistic. So with 1024 shots, the exact counts may change every time.

````python
## Let's measure now
qc = QuantumCircuit(1)

qc.h(0)

qc.measure_all()

display(qc.draw(output="mpl"))

simulator = AerSimulator()
result = simulator.run(qc, shots=1024).result()
counts = result.get_counts()

print(counts)
display(plot_histogram(counts))
````

# Phase gates

Phase gates are required and the easiest to visualize when the qubit starts from $|+\rangle$.

Why?

As show above, if we start from $|0\rangle$, many phase gates appear to do nothing on the Bloch sphere. Starting from $|+\rangle$ puts the vector on the equator, where phase rotations are visible.

````python
from qiskit.visualization.bloch import Bloch

theta = np.pi / 2
phi = np.pi / 3
lam = np.pi / 4

num_steps = 20

# ---------------------------
# Segment 1: lambda rotation
# ---------------------------
angles1 = np.linspace(0, lam, num_steps)

x1 = []
y1 = []
z1 = []

for angle in angles1:
    qc_temp = QuantumCircuit(1)
    qc_temp.h(0)
    qc_temp.rz(angle, 0)

    state_temp = Statevector.from_instruction(qc_temp)

    a = state_temp.data[0]
    b_amp = state_temp.data[1]

    x = 2 * np.real(np.conj(a) * b_amp)
    y = 2 * np.imag(np.conj(a) * b_amp)
    z = np.abs(a)**2 - np.abs(b_amp)**2

    x1.append(x)
    y1.append(y)
    z1.append(z)


# ---------------------------
# Segment 2: theta rotation
# ---------------------------
angles2 = np.linspace(0, theta, num_steps)

x2 = []
y2 = []
z2 = []

for angle in angles2:
    qc_temp = QuantumCircuit(1)
    qc_temp.rz(lam, 0)
    qc_temp.ry(angle, 0)

    state_temp = Statevector.from_instruction(qc_temp)

    a = state_temp.data[0]
    b_amp = state_temp.data[1]

    x = 2 * np.real(np.conj(a) * b_amp)
    y = 2 * np.imag(np.conj(a) * b_amp)
    z = np.abs(a)**2 - np.abs(b_amp)**2

    x2.append(x)
    y2.append(y)
    z2.append(z)


# ---------------------------
# Segment 3: phi rotation
# ---------------------------
angles3 = np.linspace(0, phi, num_steps)

x3 = []
y3 = []
z3 = []

for angle in angles3:
    qc_temp = QuantumCircuit(1)
    qc_temp.rz(lam, 0)
    qc_temp.ry(theta, 0)
    qc_temp.rz(angle, 0)

    state_temp = Statevector.from_instruction(qc_temp)

    a = state_temp.data[0]
    b_amp = state_temp.data[1]

    x = 2 * np.real(np.conj(a) * b_amp)
    y = 2 * np.imag(np.conj(a) * b_amp)
    z = np.abs(a)**2 - np.abs(b_amp)**2

    x3.append(x)
    y3.append(y)
    z3.append(z)


# ---------------------------
# Actual U gate circuit
# ---------------------------
qc = QuantumCircuit(1)

qc.u(theta, phi, lam, 0)

# display(qc.draw(output="mpl"))

state = Statevector.from_instruction(qc)
display(state.draw("latex"))

print("theta =", theta, "radians =", np.degrees(theta), "degrees")
print("phi   =", phi,   "radians =", np.degrees(phi),   "degrees")
print("lam   =", lam,   "radians =", np.degrees(lam),   "degrees")

print("red    = lambda rotation")
print("green  = theta rotation")
print("orange = phi rotation")
print("blue   = final state")


# Final Bloch vector
a = state.data[0]
b_amp = state.data[1]

xf = 2 * np.real(np.conj(a) * b_amp)
yf = 2 * np.imag(np.conj(a) * b_amp)
zf = np.abs(a)**2 - np.abs(b_amp)**2


# ---------------------------
# Bloch sphere
# ---------------------------
bloch = Bloch()

bloch.xlabel = ["x", ""]
bloch.ylabel = ["y", ""]
bloch.zlabel = ["z / |0>", "|1>"]

bloch.point_marker = ["o"]
bloch.point_size = [25]

# Add all paths first
bloch.add_points([x1, y1, z1])
bloch.add_points([x2, y2, z2])
bloch.add_points([x3, y3, z3])

# Then set colors together
bloch.point_color = ["r", "g", "orange"]

# Final blue arrow
bloch.vector_color = ["b"]
bloch.add_vectors([xf, yf, zf])

bloch.show()
````

# Two-qubit gates

## CNOT / CX gate

The `CX` gate has:

```text
control qubit
target qubit
```

Simple rule:

```text
If control = 1, flip target.
If control = 0, do nothing to target.
```

In Qiskit:

```python
qc.cx(control, target)
```

````python
qc = QuantumCircuit(2)

qc.x(0)        # make q0 = 1

# applies CNOT: if q0 is 1, flip q1
# control qubit = q0
# target qubit  = q1
qc.cx(0, 1)

display(qc.draw(output="mpl"))

state = Statevector.from_instruction(qc)

display(state.draw("latex"))

qc.measure_all()

display(qc.draw(output="mpl"))

simulator = AerSimulator()
result = simulator.run(qc, shots=1024).result()
counts = result.get_counts()

print(counts)
display(plot_histogram(counts))

## moral of the story is: CNOT flips the second qubit only when the first qubit is 1.
````

## CZ gate

The `CZ` gate is a controlled phase gate.

Simple idea:

```text
If both qubits are 1, add a phase of -1.
```

Its effect may not always be visible in direct measurement, because measurement only shows probabilities, not phase.

````python
# create 2 qubits
qc = QuantumCircuit(2)

# apply Hadamard on both qubits.
qc.h(0)
qc.h(1)

# now both become superposition
# q0 = (|0⟩ + |1⟩) / √2
# q1 = (|0⟩ + |1⟩) / √2
# in total the state becomes: (|00⟩ + |01⟩ + |10⟩ + |11⟩) / 2

# apply, controlled-z gate... ff both qubits are 1, add minus (phase) sign to |11⟩.
qc.cz(0, 1)

display(qc.draw(output="mpl"))
state = Statevector.from_instruction(qc)
display(state.draw("latex"))

## moral of the story: CZ does not flip the qubit.
# CZ adds a phase/sign change when both qubits are 1.
````

# Entanglement

Entanglement is a special quantum connection between qubits.

A simple two-qubit entangled state is the Bell state:

$$
\frac{|00\rangle + |11\rangle}{\sqrt{2}}
$$

Simple meaning:

- The result is not fixed before measurement.
- But when measured, both qubits match.
- We get mostly `00` and `11`, not `01` and `10`.

## Bell state circuit

Steps:

1. Apply `H` on qubit 0.
2. Apply `CX` from qubit 0 to qubit 1.

````python
# lets create 2 qubits
    # q0 = |0⟩
    # q1 = |0⟩
# so now the initial state is :
# |00⟩
qc = QuantumCircuit(2)

# Apply Hadamard gate on q0
qc.h(0)
# now q0 goes into superposition: q0 = (|0⟩ + |1⟩) / √2
# and q1 = |0⟩
# their combined state is:
# (|00⟩ + |10⟩) / √2

# this line applies CNOT gate
qc.cx(0, 1)
# we know what is does: if q0 is 1, flip q1
# so:
# |00⟩ stays |00⟩
# |10⟩ becomes |11⟩
# now the final "Bell" state is:
# (|00⟩ + |11⟩) / √2

# '''
# moral of the stori:
# The two qubits are now linked. CNOT gate links the first second withe first qubit

# If q0 is measured as 0, q1 will also be 0.
# If q0 is measured as 1, q1 will also be 1.

# This is Entanglement

# So the qubits are not random independently.
# --> They are correlated.

# It also means the information is stored in the joint two-qubit state, not in each qubit alone.
# '''

display(qc.draw(output="mpl"))
state = Statevector.from_instruction(qc)
# display(state.draw("latex"))
````

````python
# let's add the measurement code
qc = QuantumCircuit(2)

qc.h(0)
qc.cx(0, 1)

qc.measure_all()

display(qc.draw(output="mpl"))

simulator = AerSimulator()
result = simulator.run(qc, shots=1024).result()
counts = result.get_counts()

print(counts)
display(plot_histogram(counts))

````

## Quantu coin toss circuit

1. Quantum Coin Toss using Hadamard Gate
2. Initial state = |0>
3. Apply H gate
4. Final state = (|0> + |1>) / sqrt(2)
5. Expected measurement result:
    * about 50% |0> and 50% |1>


````python
qc = QuantumCircuit(1)

print("Step 1: Initial circuit")
display(qc.draw(output="mpl"))

state = Statevector.from_instruction(qc)

print("Initial statevector:")
display(state.draw("latex"))

print("Initial Bloch sphere:")
display(plot_bloch_multivector(state))
````

````python
## here is the Hadamard matrix

H = (1 / np.sqrt(2)) * np.array([[1, 1],
                                 [1, -1]])

print("Hadamard matrix H =")
print(H)
'''
This is the matrix form of the Hadamard gate:
H = (1/√2) [ [1, 1],
             [1,-1] ]

It converts |0⟩ into equal superposition.


'''
````

````python
## Now let's apply Hadamard gate
qc = QuantumCircuit(1)

qc.h(0)

print("Step 2: After applying H gate")
display(qc.draw(output="mpl"))

state = Statevector.from_instruction(qc)

print("Final statevector:")
display(state.draw("latex"))

print("Bloch sphere after H gate:")
display(plot_bloch_multivector(state))
'''
qc.h(0) means apply H gate on qubit 0

Final state becomes:
(|0⟩ + |1⟩) / √2

On Bloch sphere, the arrow moves from the top to the equator - This shows the qubit is now in superposition

'''
````

````python
## show probabilities before meeasurements
qc = QuantumCircuit(1)

qc.h(0)

state = Statevector.from_instruction(qc)

print("Statevector:")
display(state.draw("latex"))

probs = state.probabilities_dict()

print("theoretical probabilities:")
print(probs)
````

````python
## Let's run this cicuit and measure
qc = QuantumCircuit(1)

qc.h(0)
qc.measure_all()

print("Step 3: Circuit with measurement")
display(qc.draw(output="mpl"))

simulator = AerSimulator()

# shots=1024 means run the same circuit 1024 times
result = simulator.run(qc, shots=1024).result()
counts = result.get_counts()

print("Measurement counts:")
print(counts)

display(plot_histogram(counts))
````

### Conclusion:
1. We start with |0⟩.
2. Then we apply Hadamard gate.
3. Hadamard creates an equal superposition of |0⟩ and |1⟩.
4. So when we measure many times, we get about 50% 0 and 50% 1.
5. That is why this acts like a quantum coin toss.

## Double Hadamard Gate

|0⟩ -----> H -----> H -----> |0⟩

**Main idea:**

* Hadamard gate creates superposition.

* But if we apply Hadamard twice, it brings the qubit back to the original state.

* H followed by H = no change

**Mathematically:**

    H² = I

So:

    H|0⟩ = (|0⟩ + |1⟩)/√2

    H(H|0⟩) = |0⟩


This shows reversibility.


````python
## Let's create this circuit:
qc = QuantumCircuit(1)

qc.h(0)
qc.h(0)

print("Full Double Hadamard circuit:")
display(qc.draw(output="mpl"))
````

````python
## visualising the initial state:
qc = QuantumCircuit(1)

print("Step 0: Initial state")

display(qc.draw(output="mpl"))

state = Statevector.from_instruction(qc)

print("Statevector:")
display(state.draw("latex"))

print("Bloch sphere:")
display(plot_bloch_multivector(state))
````

````python
## After applying the first Hadamard

qc = QuantumCircuit(1)

qc.h(0)

print("Step 1: After first H gate")

display(qc.draw(output="mpl"))

state = Statevector.from_instruction(qc)

print("Statevector:")
display(state.draw("latex"))

print("Bloch sphere:")
display(plot_bloch_multivector(state))

'''
First H gate changes |0> into superposition.

|0> becomes:

(|0> + |1>) / sqrt(2)

Now the Bloch sphere arrow moves to the equator.

'''
````

````python
## now after applying second Hadamard:

qc = QuantumCircuit(1)

qc.h(0)
qc.h(0)

print("Step 2: After second H gate")

display(qc.draw(output="mpl"))

state = Statevector.from_instruction(qc)

print("Statevector:")
display(state.draw("latex"))

print("Bloch sphere:")
display(plot_bloch_multivector(state))

'''
Second H gate reverses the first H gate.

The qubit comes back to |0>.

Bloch sphere arrow goes back to the top.

'''
````

````python
## let's see the probabilities
qc = QuantumCircuit(1)

qc.h(0)
qc.h(0)

state = Statevector.from_instruction(qc)

print("Final statevector:")
display(state.draw("latex"))

probs = state.probabilities_dict()

print("Final probabilities:")
print(probs)

'''
means:
100% chance of measuring 0.
0% chance of measuring 1.

'''
````

````python
## and if we run and measure, we get:

qc = QuantumCircuit(1)

qc.h(0)
qc.h(0)

qc.measure_all()

print("Circuit with measurement:")
display(qc.draw(output="mpl"))

simulator = AerSimulator()
result = simulator.run(qc, shots=1024).result()
counts = result.get_counts()

print("Measurement counts:")
print(counts)

display(plot_histogram(counts))

'''
means:
All measurements give 0.

'''
````

### Conclusion

We start with |0>

First H gate creates a 50-50 superposition:

    (|0> + |1>) / sqrt(2)

Second H gate reverses this operation.

So the final state becomes |0> again.

This proves that H is reversible.

Applying H twice is the same as doing nothing.


*IMPORTANT NOTE:*

* One H gives randomness during measurement
* But two H gates removes that randomness and return the qubit to |0>.



## Quantum Random Number Generator

**Main idea**

We use the same idea as quantum coin toss.

    |0⟩ ---- H ---- Measurement

Hadamard creates:

    (|0⟩ + |1⟩) / sqrt(2)

So every measurement gives:

    0 or 1 randomly

If we repeat this many times, we get a random binary string like:

    101001011010...

````python
## the circuit for this

qc = QuantumCircuit(1)

qc.h(0)
qc.measure_all()

print("Quantum random bit circuit:")
display(qc.draw(output="mpl"))

'''
H gate creates 50-50 superposition.

and measurement converts it into either 0 or 1.

'''
````

````python
## let's check the state BEFORE measurements
qc = QuantumCircuit(1)

qc.h(0)

print("State before measurement:")

display(qc.draw(output="mpl"))

state = Statevector.from_instruction(qc)

print("Statevector:")
display(state.draw("latex"))

print("Bloch sphere:")
display(plot_bloch_multivector(state))

print("Probabilities:")
print(state.probabilities_dict())
````

````python
## let's build the random num generator circuit, run and measure it

N = 32 ## seed or shots val

qc = QuantumCircuit(1)

qc.h(0)
qc.measure_all()

simulator = AerSimulator()

# shots (does the trick) = N var means we repeat the circuit N times.
# memory=True, it means store every single measurement result
# we then join all bits to make one binary string. Simple!
result = simulator.run(qc, shots=N, memory=True).result()
bits = result.get_memory()
binary_string = "".join(bits)

print("Generated random bits:")
print(bits)

print()
print("Generated binary string:")
print(binary_string)
````

````python
## let's derrive its frequency stats, for fun!
N = 1024

qc = QuantumCircuit(1)

qc.h(0)
qc.measure_all()

simulator = AerSimulator()

result = simulator.run(qc, shots=N, memory=True).result()
bits = result.get_memory()
binary_string = "".join(bits)

count_0 = binary_string.count("0")
count_1 = binary_string.count("1")

percent_0 = (count_0 / N) * 100
percent_1 = (count_1 / N) * 100

print("Total bits generated =", N)
print("Number of 0s =", count_0)
print("Number of 1s =", count_1)

print("Percentage of 0s =", percent_0)
print("Percentage of 1s =", percent_1)
````

````python
counts = result.get_counts()

print("Counts:")
print(counts)

display(plot_histogram(counts))
````

### Conclusion


> `H + Measure = random bit`

> `Repeat N times = random binary string`



## Simple Grover Search (2 qubits)

**Main idea**

Grover search has 3 main parts:

1. Create equal superposition

    so all states are equally possible

2. Oracle

    mark the target state

3. Diffuser

    increase the probability of the target state

For 2 qubits, the possible states are:

    |00⟩, |01⟩, |10⟩, |11⟩

At first, all 4 states have equal chance:

    25% each

After Grover amplification, the target state |11⟩ becomes much more likely.

    For this very small 2-qubit example, after one Grover iteration, we get:

    |11⟩ with almost 100% probability

````python
## our circuit
qc = QuantumCircuit(2)

# Create Equal Superposition
qc.h(0)
qc.h(1)

# Oracle to set the target for |11>
qc.cz(0, 1)

# Diffuser, for amplification
qc.h(0)
qc.h(1)

qc.x(0)
qc.x(1)

qc.cz(0, 1)

qc.x(0)
qc.x(1)

qc.h(0)
qc.h(1)

print("Full Grover circuit:")
display(qc.draw(output="mpl"))
````

### Let's go step by step

````python
## check the init state
qc = QuantumCircuit(2)

print("Step 0: Initial state")

display(qc.draw(output="mpl"))

state = Statevector.from_instruction(qc)

print("Statevector:")
display(state.draw("latex"))

print("Probabilities:")
print(state.probabilities_dict())

print("Bloch sphere:")
display(plot_bloch_multivector(state))

'''
We start with |00>.
Only |00> has probability 1.
All other states have probability 0. Because that's the default

'''
````

````python
## let's try to create Equal Superposition
qc = QuantumCircuit(2)

qc.h(0)
qc.h(1)

print("Step 1: After applying H on both qubits")

display(qc.draw(output="mpl"))

state = Statevector.from_instruction(qc)

print("Statevector:")
display(state.draw("latex"))

print("Probabilities:")
print(state.probabilities_dict())

print("Bloch sphere:")
display(plot_bloch_multivector(state))
````

After Hadamard on both qubits:

    |00> becomes... (|00> + |01> + |10> + |11>) / 2

Now all 4 states are equally likely:

    00 = 25%
    01 = 25%
    10 = 25%
    11 = 25%

till now, we have NOT searched anything.


````python
## calling the Oracle, for setting the target as |11>
qc = QuantumCircuit(2)

qc.h(0)
qc.h(1)

qc.cz(0, 1)

print("Step 2: After oracle")

display(qc.draw(output="mpl"))

state = Statevector.from_instruction(qc)

print("Statevector:")
display(state.draw("latex"))

print("Probabilities:")
print(state.probabilities_dict())

# print("Bloch sphere:")
# display(plot_bloch_multivector(state))

# '''
# It only marks the target using phase sign!
# no the state is:
# (|00> + |01> + |10> - |11>) / 2

# '''
````

````python
## constructing the diffuser
qc = QuantumCircuit(2)

qc.h(0)
qc.h(1)

qc.cz(0, 1)

qc.h(0)
qc.h(1)

qc.x(0)
qc.x(1)

qc.cz(0, 1)

qc.x(0)
qc.x(1)

qc.h(0)
qc.h(1)

print("Step 3: After diffuser")

display(qc.draw(output="mpl"))

state = Statevector.from_instruction(qc)

print("Statevector:")
display(state.draw("latex"))

print("Probabilities:")
print(state.probabilities_dict())

print("Bloch sphere:")
display(plot_bloch_multivector(state))
````

The diffuser is the part that amplifies the marked state.

After the diffuser, the probability moves toward the target state.

For this 2-qubit case, after one Grover iteration, the final state becomes:

    |11>

So now:

    P(11) = ~1

and the other states become:

    P(00) = ~0
    P(01) = ~0
    P(10) = ~0

This is the main Grover effect:

    mark target first,
    then amplify it

````python
## adding measurements code
qc = QuantumCircuit(2)

qc.h(0)
qc.h(1)

qc.cz(0, 1)

qc.h(0)
qc.h(1)

qc.x(0)
qc.x(1)

qc.cz(0, 1)

qc.x(0)
qc.x(1)

qc.h(0)
qc.h(1)

qc.measure_all()

print("Grover circuit with measurement:")
display(qc.draw(output="mpl"))

simulator = AerSimulator()
result = simulator.run(qc, shots=1024).result()
counts = result.get_counts()

print("Measurement counts:")
print(counts)

display(plot_histogram(counts))
````

This histogram shows that the circuit found the target state:

    |11>

almost every time.

**That is the goal of Grover search.**

**Further explanation**

**Before Grover:**

    all states are equally likely

**After Grover:**

    target state becomes the most likely

> That is why Grover is called a *search algorithm*.
